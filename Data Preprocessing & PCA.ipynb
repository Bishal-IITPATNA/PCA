```python
#Data Pre-processing Standardization Of Data
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
data=d.head(15000)
print(data.head(5))
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data)
```

```python
print(scaled_data.shape)
```

```python
sample_data=scaled_data
```

```python
cov_matrix=np.matmul(sample_data.T,sample_data)
```

```python
print(cov_matrix.shape)
```

```python
print(np.mean(sample_data,axis=0))
```

```python
#library to import eigen value function
from scipy.linalg import eigh
```

```python
#The parameter eighvals returns eigen value from low to high
#eigh function generates eigen value in ascending order
#This code returns top 2 eigen value i.e, 783 and 784
values,vectors=eigh(cov_matrix,eigvals=(782,783))
values1,vectors1=eigh(cov_matrix)
```

```python
print(values1)
```

```python
print(values)
```

```python
print(vectors.shape)
```

```python
print("percentage of variance explained",sum(values)/sum(values1))
```
